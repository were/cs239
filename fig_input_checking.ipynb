```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

from __future__ import division
from __future__ import print_function
from __future__ import absolute_import
from __future__ import unicode_literals  

import numpy as np
import matplotlib.pyplot as plt
```

```python
iduti= np.load('./output/iduti_results.npz')
duti= np.load('./output/duti_results.npz')
```

```python
fig, axs = plt.subplots(1, 1, figsize=(5.5,3.5))

axs.plot(iduti['ratio'], iduti['corrected'], label='Our iDUTI flow')
axs.plot(duti['ratio'], duti['corrected'], label='Optimized DUTI')
axs.plot(duti['ratio'], duti['ratio'], linestyle=':', label='Traditional Optimal')
axs.set_xlim([0, 50])
axs.legend(loc=(0.58,0.02))
axs.set_xlabel('# of data inspected')
axs.set_ylabel('# of data corrected')

plt.tight_layout()
plt.subplots_adjust(wspace=0.28)
```
